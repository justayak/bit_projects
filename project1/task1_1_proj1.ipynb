```python
#Solution:
#filter to remove outliers
fliter_WO1 = np.array([True if (x > 0) else False for x in np.nditer(X[0:1])])
fliter_WO2 = np.array([True if (x > 0) else False for x in np.nditer(X[1:2])])
fliter_WO = np.logical_and(fliter_WO1, fliter_WO2)

X_WO = X[:, fliter_WO]
whExample.plotData2D(X_WO)
```

```python
# next, let's plot height vs. weight 
# first, copy information rows of X into 1D arrays
w_WO = np.copy(X_WO[0, :])
h_WO = np.copy(X_WO[1, :])

# second, create new data matrix Z by stacking h and w
Z = np.vstack((h_WO, w_WO))
# third, plot this new representation of the data
#plotData2D(Z, 'plotHW.pdf')
whExample.plotData2D(Z)
```
